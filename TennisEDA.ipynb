```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

stat2010 = pd.read_csv("TennisProject/match_stats_2010.csv")
rank2010 = pd.read_csv("TennisProject/atp_matches_2010.csv")
```

```python
partStat = stat2010.iloc[1870:2223]
partrank2010 = rank2010
```

```python
#Getting rid of unwanted columns
partStat.drop(["match_id", "match_stats_url_suffix","match_time","winner_slug","loser_slug"], inplace = True, axis = 1)
partStat
```

```python
#look at how serves overall effect winning
graph = sns.scatterplot(x= partStat["loser_serve_rating"], y = partStat["winner_serve_rating"])
graph.set(xlabel = "Loser Serve Rating", ylabel = "Winner Serve Rating", title = "Winner Serve Rating vs Loser Serve Rating")
plt.ylim(100, 400)
plt.xlim(100, 400)
```

```python
#Break down what matters most in serving
#Aces first
dfAces = partStat[["loser_aces","winner_aces"]]
graph = sns.scatterplot(x= dfAces["loser_aces"], y = dfAces["winner_aces"])
graph.set(xlabel = "Loser Aces", ylabel = "Winner Aces", title = "Winner Aces vs Loser Aces")
```

```python
dfFaults = partStat[["loser_double_faults","winner_double_faults"]]
graph = sns.scatterplot(x = dfFaults["loser_double_faults"],y=dfFaults["winner_double_faults"])
graph.set(xlabel = "Loser Double Faults", ylabel = "Winner Double Faults", title = "Winner Faults vs Loser Faults")
plt.xlim(0,14)
plt.ylim(0,14)
```

```python
graph = sns.scatterplot(x = partrank2010["loser_rank"],y=partrank2010["winner_rank"])
graph.set(xlabel = "Loser ATP Rank", ylabel = "Winner ATP rank", title = "Winner Rank vs Loser Rank")
plt.xlim(0,400)
plt.ylim(0,400)
```

```python
#Does first serve percentage matter
wfirstServePer = []
lfirstServePer= []
for i in range(353):
    wfirstServePer.append(((partStat.at[(i+1870),"winner_first_serves_in"])/(partStat.at[(i+1870),"winner_first_serves_total"])))
    lfirstServePer.append(((partStat.at[(i+1870),"loser_first_serves_in"])/(partStat.at[(i+1870),"loser_first_serves_total"])))
graph = sns.scatterplot(x = lfirstServePer , y = wfirstServePer)
graph.set(xlabel = "Loser 1st Serve Percentage", ylabel = "Winner 1st serve percentage", title = "Winner 1st % vs Loser 1st %")
plt.ylim(0.2,1)
plt.xlim(0.2,1)
```

```python
wServeBreakPer = []
lServeBreakPer= []
for i in range(353):
    wServeBreakPer.append(((partStat.at[(i+1870),"winner_break_points_saved"])/(partStat.at[(i+1870),"winner_break_points_serve_total"])))
    lServeBreakPer.append(((partStat.at[(i+1870),"loser_break_points_saved"])/(partStat.at[(i+1870),"loser_break_points_serve_total"])))
graph = sns.scatterplot(x = lServeBreakPer , y = wServeBreakPer)
graph.set(xlabel = "Loser Break Points Saved Percentage", ylabel = "Winner Break Points Saved percentage", title = "Winner Break % vs Loser Break %")
plt.ylim(0,1.2)
plt.xlim(0,1.2)
```
